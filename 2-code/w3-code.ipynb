```python
def analiza(data,route):

    print('analizo los datos de ',route)

    shape_ori=data.shape
    shape_duplica=data.drop_duplicates().shape
    
    print('Shape',shape_ori, '\nDatos duplicados',shape_ori!=shape_duplica,'\n')

    '''for c in data.select_dtypes('integer').columns:
        data[c] = pd.to_numeric(data[c], downcast='integer')'''

    if 'last_update' in data.columns:

        data.last_update = pd.to_datetime(data.last_update)

    print(data.info())
   
    return data.head()
```

```python
def duplos(data):

    longitud=len(data)
    unicos=len(data.unique())

    if longitud==unicos:
        return 'Todo OK'
    else:
        return 'Hay alguno repetido'

```
